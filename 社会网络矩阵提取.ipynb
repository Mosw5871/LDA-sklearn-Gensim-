```python
import pandas as pd
```

```python
# 读取CSV文件
data = pd.read_csv("/Users/shiwei/Desktop/selected_data.csv")
```

```python
data
```

```python
# 去除NaN值
data = data.dropna(subset=['entities.mentions'])
```

```python
data
```

```python
# 重置索引以解决索引问题
data.reset_index(drop=True, inplace=True)
data
```

```python
data['entities.mentions'][1]
```

```python
# 去除方括号和引号中的空格，然后用逗号分隔用户名
data['entities.mentions'] = data['entities.mentions'].str.replace(r'[\[\]"\s@]', '', regex=True)
```

```python
data
```

```python
# 将entities.mentions中的多个用户名分拆为多行
data['entities.mentions'] = data['entities.mentions'].str.split(',')

# 使用explode函数将多个用户名分拆为多行，并与原始数据一一对应
data = data.explode('entities.mentions')

# 重置索引以解决索引问题
data.reset_index(drop=True, inplace=True)
```

```python
data
```
