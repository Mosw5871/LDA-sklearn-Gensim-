```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv("E:\一带一路数据分析\345家媒体全部帖子16493.csv", encoding="utf-8", index_col=[0])
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
sentiment_list = []
for sentence in df["text"]:
    vs = analyzer.polarity_scores(sentence)
    sentiment_list.append(vs)
    # print("{:-<65} {}".format(sentence, str(vs)))
```

```python
sentiment_list[0:5]
```

```python
df_sentiment_list = pd.DataFrame(sentiment_list)
df_sentiment_list
```

```python
df_cov = pd.concat([df, df_sentiment_list], axis=1)
```

```python
df_cov.to_csv(r"E:\一带一路数据分析\345家媒体全部帖子16493情感分析结果.csv", encoding="utf-8")
```

```python
df_cov_senti = df_cov[["author.username", "neg", "neu", "pos", "compound"]]
```

```python
df_cov_senti
```

```python
# 计算平均情感值
df_mean_senti = df_cov_senti.groupby("author.username").agg("mean")
```

```python
df_mean_senti.to_csv("E:\\一带一路数据分析\\345家媒体平均情感.csv", encoding="utf-8")
```
